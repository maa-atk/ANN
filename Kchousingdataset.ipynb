**Feature Engineering**

remove id

year and month extracted from the date parameter 

```
df=df.drop('id',axis=1)
```

```
df['date']=pd.to_datetime(df['date'])
#converted it to splittable objects
df['year']=df['date'].apply(lambda date:date.year) 

df['month']=df['date'].apply(lambda date:date.month) 
df=df.drop(['date'],axis=1)
```

Model making

```
X=df.drop('price',axis=1).values
y=df['price'].values
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```
X_train=scaler.fit_transform(X_train)
```

```
X_test=scaler.transform(X_test)
#only transform the test set
#no prior information
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```
#convention is layers of neurons== no of incoming feature data

model=Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))


#final layer
model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=128,epochs=400)
#smaller the batches more the time to train less over fit
#batches in pow of size
```

```
losses=pd.DataFrame(model.history.history)
losses.plot()
#could keep training no spikes found
```

```
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```
predictions = model.predict(X_test)
```

```
np.sqrt(mean_squared_error( y_test,predictions))
```

```
mean_absolute_error(y_test,predictions)
```
